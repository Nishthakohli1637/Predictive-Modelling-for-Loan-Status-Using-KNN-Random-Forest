```python
# removing outliers

# filtering the list of specific columns
col = ['CNT_CHILDREN', 'AMT_INCOME_TOTAL' , 'CNT_FAM_MEMBERS']

# for loop to remove outliers
for i in range(len(col)):
    # calculating the first and third quartile
    q1 = df[col[i]].quantile(0.25)
    q3 = df[col[i]].quantile(0.75)
    # calculating the interquartile range
    iqr = q3 - q1
    # calculating the lower and upper bounds
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    # replacing the outliers with the median
    df[col[i]] = np.where((df[col[i]] >= upper_bound) | (df[col[i]] <= lower_bound), df[col[i]].median() , df[col[i]])
```

```python
# X=scaled_feature.drop(['STATUS', 'CNT_CHILDREN', 'FLAG_MOBIL', 'CNT_FAM_MEMBERS'], axis=1)
X=df_feat.drop(['CNT_CHILDREN', 'FLAG_MOBIL', 'CNT_FAM_MEMBERS'], axis=1)
```

```python
data.head()
```

```python
y=data['STATUS']
```

```python
data.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=7)
```

```python
knn.fit(x_train, y_train)
```

```python
pred=knn.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
pd.DataFrame({'Actual':y_test, 'Predicted':pred})
```

```python
error_rate=[]
for i in range(1,10):
        knn=KNeighborsClassifier(n_neighbors=i)
        knn.fit(x_train,y_train)
        pred_i=knn.predict(x_test)
        error_rate.append(np.mean(pred_i!=y_test))
```

```python
plt.plot(range(1,10), error_rate, color='blue', linestyle='dashed', marker='o', markerfacecolor='red')
plt.title('Error Rate vs K-Value')
plt.xlabel('K')
plt.ylabel('Error_Rate')
```
